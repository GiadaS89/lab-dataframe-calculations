#### Cleanup `Name` that contain "Mega".

If you have checked out the pokemon names carefully enough, you should have found there are junk texts in the pokemon names which contain "Mega". We want to clean up the pokemon names. For instance, "VenusaurMega Venusaur" should be "Mega Venusaur", and "CharizardMega Charizard X" should be "Mega Charizard X".

```python
# your code here

#Premise: Localize where the word 'Mega'is in the column Name, in order to retrive from the dataframe pokemon, which values or strings contain the word 'Mega'

pokemon.loc[pokemon['Name'].str.contains('Mega', case=False)]

```

```python
#1: We understand that 'Mega' is a substring of the strings "VenusaurMega Venusaur" and "CharizardMega Charizard X", which appear in the column Name from the dataframe
# pokemon. We apply the find() specifically to both our strings to understand from which position/index our substring Mega starts. 

## Case for VenusaurMega Venusaur

name = 'VenusaurMega Venusaur' #string
mega = 'Mega' #substring of string 'mega'

print(name.find(mega))

start_index_of_mega = name.find(mega)

new_name = name[start_index_of_mega:]
new_name
```

```python
## Case for CharizardMega Charizard X

name = 'CharizardMega Charizard X' #string
mega = 'Mega' #substring of string 'mega'

print(name.find(mega))

start_index_of_mega = name.find(mega)

new_name = name[start_index_of_mega:]
new_name
```

```python
#2: We realize that in above cases, our variable or substring mega starts in position 8 for Mega Venusaur whereas in position 9 for string CharizardMega Charizard X. 
# We then create a generic function to clean the names/strings in scope and return the cleaned_name in case 'mega', as a substring, is found in the string. 

mega = 'Mega' #substring of whatever string we find in column Name

def clean_name(name):
    if 'Mega'in name:
        start_index_of_mega = name.find(mega)
        cleaned_name = name[start_index_of_mega:]    
        return cleaned_name
    else:
        return name
```

```python
#3. Then we apply our generic function - apply(clean_name) - to the column 'Name'of our dataframe pokemon, and generate a new column called 'clean_name' where
#the cleaned names will be rendered. 

pokemon['clean_name'] = pokemon['Name'].apply(clean_name)
pokemon
```

#### Create a new column called `Combo Type` whose value combines `Type 1` and `Type 2`.

Rules:

* If both `Type 1` and `Type 2` have valid values, the `Combo Type` value should contain both values in the form of `<Type 1> <Type 2>`. For example, if `Type 1` value is `Grass` and `Type 2` value is `Poison`, `Combo Type` will be `Grass-Poison`.

* If `Type 1` has valid value but `Type 2` is not, `Combo Type` will be the same as `Type 1`. For example, if `Type 1` is `Fire` whereas `Type 2` is `NaN`, `Combo Type` will be `Fire`.

```python
#1. Replace the NaN values for the entire pokemon dataframe, and reassign it again to the dataframe pokemon because .fillna() is not an inplace method

pokemon = pokemon.fillna(' ')

#We can also replace the NaN values only for column Type 2 --> pokemon['Type 2'] = pokemon['Type 2'].fillna(' ')
```
